```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv('/Users/srilathajuvvadi/Library/Containers/com.microsoft.Excel/Data/Desktop/movies_DATASET.csv')
df.head()
```

```python
#to check for any missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#to get the datatypes of the columns
df.dtypes
```

```python
#sorting data by thier gross values(higher to lower)
df.sort_values(by=['gross'],inplace=False,ascending=False)
df.head()
```

```python
#preview of full data
pd.set_option('display.max_rows',None)
```

```python
#to show if any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
df.head()
```

```python
# comparison between gross and budget 
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs gross earnings')
plt.xlabel('gross earnings')
plt.ylabel('budget for film')
plt.show()
```

```python
#budget vs gross using seaborn
sns.regplot(x="budget", y="gross", data=df,scatter_kws={"color":"black"},line_kws={"color":"purple"})
```

```python
#start looking at correlation
```

```python
df.corr(method ='pearson',numeric_only=True)
```

```python
df.corr(method ='kendall',numeric_only=True)
```

```python
#correlation matrix
correlation_matrix = df.corr(method ='pearson',numeric_only=True)

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python

sorted_pairs=corr_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
